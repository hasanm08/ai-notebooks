```python
import pandas as pd
dataset_dfs={
    "Breast":pd.read_csv('/content/BreastTissue.txt', sep="\t", header=None),
    "Diabetes":pd.read_csv('/content/Diabetes.txt', sep="\t", header=None).drop(columns=[9]), 
    "Glass":pd.read_csv('/content/Glass.txt', sep="\t", header=None).drop(columns=[10]),
    "Ionosphere":pd.read_csv('/content/Ionosphere.txt', sep=",", header=None).drop(columns=[1]),
    "Sonar":pd.read_csv('/content/Sonar.txt', sep=",", header=None),
    "Wine":pd.read_csv('/content/Wine.txt', sep=",", header=None),
}
```

```python
def normalize(df):
  #brings all antecedants in [0,100] interval 
    result = df.copy()
    for feature_name in result.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name]=((df[feature_name] - min_value) / (max_value - min_value))*10
    return result
```

```python
dataset_input={
    "Breast":normalize(dataset_dfs['Breast'].iloc[:,:-1]),
    "Diabetes":normalize(dataset_dfs['Diabetes'].iloc[:,:-1]), 
    "Glass":normalize(dataset_dfs['Glass'].iloc[:,:-1]),
    "Ionosphere":normalize(dataset_dfs['Ionosphere'].iloc[:,:-1]),
    "Sonar":normalize(dataset_dfs['Sonar'].iloc[:,:-1]),
    "Wine":normalize(dataset_dfs['Wine'].iloc[:,:-1]),
}
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le2 = preprocessing.LabelEncoder()

dataset_label={
    "Breast":dataset_dfs['Breast'].iloc[:,-1],
    "Diabetes":dataset_dfs['Diabetes'].iloc[:,-1], 
    "Glass":dataset_dfs['Glass'].iloc[:,-1],
    "Ionosphere":le.fit_transform(dataset_dfs['Ionosphere'].iloc[:,-1]),
    "Sonar":le2.fit_transform(dataset_dfs['Sonar'].iloc[:,-1]),
    "Wine":dataset_dfs['Wine'].iloc[:,-1],
}
```

```python
#Importing the required modules
from sklearn.preprocessing import StandardScaler
from math import sqrt
import numpy as np
import time
from scipy.stats import mode
from sklearn.model_selection import train_test_split
from sklearn.metrics import pairwise,accuracy_score,f1_score
class KernelizedKNN:
  def __init__(self,x,y,kernel):
    self.X_train, self.X_test, self.y_train, self.y_test =train_test_split(np.array(x),np.array(y), train_size=.7, random_state=8, shuffle=True)
    self.k=1
    # Feature Scaling
    sc = StandardScaler()
    self.X_train = sc.fit_transform(self.X_train)
    self.X_test = sc.transform(self.X_test)
    #kernel
    if(kernel=="p1"):
      self.X_train=pairwise.polynomial_kernel(self.X_train, degree=1, coef0=1)
    elif(kernel=="p2"):
      self.X_train=pairwise.polynomial_kernel(self.X_train, degree=2, coef0=1)
    elif(kernel=="p3"):
      self.X_train=pairwise.polynomial_kernel(self.X_train, degree=3, coef0=1)
    elif(kernel=="rbf"):
      for index in range(0,100,2):
        gamma=index/10
        max_score={
            "accuracy":0,
            "f1":0,
            "time":0,
            "gamma":0
        }
        acc_gamma=()
        self.X_train=pairwise.rbf_kernel(self.X_train, gamma=gamma)
        # print(f"-----------------------1NN with Kernel: {kernel} gamma: {gamma}-----------------------")
        start=time.time()
        yp=self.predict(self.X_test)
        end=time.time()
        elapsed_time = end - start
        # print('Execution time:', elapsed_time, 'seconds')
        acc=accuracy_score(self.y_test,yp)
        f1=f1_score(self.y_test,yp,average="macro")
        # print('accuracy:',acc)
        # print('f1 score:',f1)
        if(acc>max_score['accuracy']):
          max_score={
            "accuracy":acc,
            "f1":f1,
            "time":elapsed_time,
            "gamma":gamma }
      print(f"-----------------------1NN with RBF Kernel best case -----------------------")
      print(max_score)
      print()
      return
      pass
    elif(kernel=="linear"):
      self.X_train=pairwise.linear_kernel(self.X_train)
      pass
    elif(kernel==None):
      pass
    print(f"-----------------------1NN with Kernel: {kernel}-----------------------")
    start=time.time()
    yp=self.predict(self.X_test)
    end=time.time()
    elapsed_time = end - start
    print('Execution time:', elapsed_time, 'seconds')
    acc=accuracy_score(self.y_test,yp)
    f1=f1_score(self.y_test,yp,average="macro")
    print('accuracy:',acc)
    print('f1 score:',f1)
    print()

  def calculate_euclidean(self,sample1,sample2):
    distance=0.0
    for i in range(len(sample1) if len(sample1)<len(sample2)else len(sample2)):
      # print(sample1)
      # print(sample2)
      distance+=(sample1[i]-sample2[i])**2 #Euclidean Distance = sqrt(sum i to N (x1_i – x2_i)^2)
    return sqrt(distance)
  def nearest_neighbors(self,test_sample):
    distances=[]#calculate distances from a test sample to every sample in a training set
    for i in range(len(self.X_train)):
      distances.append((self.y_train[i],self.calculate_euclidean(self.X_train[i],test_sample)))
    distances.sort(key=lambda x:x[1])#sort in ascending order, based on a distance value
    neighbors=[]
    for i in range(self.k): #get first k samples
      neighbors.append(distances[i][0])
    return neighbors
  def predict(self,test_set):
    predictions=[]
    for test_sample in test_set:
      neighbors=self.nearest_neighbors(test_sample)
      labels=[sample for sample in neighbors]
      prediction=max(labels,key=labels.count)
      predictions.append(prediction)
    return predictions
```
