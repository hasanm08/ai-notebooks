#Init

```python
import random
#initialization
def init_gen():
  gen=[]
  i=1
  while len(gen)<8:
    x=random.randint(1,8)
    if x not in [item[1] for item in gen]:
      gen.append((i,x))
      i=i+1
  return gen
```

```python
def init_population(count):
  population=[]
  for i in range(0,count):
    population.append(init_gen())
  
  return population
```

#Fitness

```python
#fitness
def fitness(gen):
  error=0
  clone=gen.copy()
  for i in range(0,len(clone)):
    x1,y1=clone[i]
    for j in range(i+1,len(clone)):
      x2,y2=clone[j]
      if x1==x2:
        error= error+1
      if y1==y2:
        error= error+1
      if x1-x2==-(y1-y2):
        error= error+1
      if x1-x2==y1-y2:
        error= error+1
  return error
```

#Parent selection

```python
def key(e):
  return e[1]
def parent_selection(population):
  random.shuffle(population)
  fitness_list=[(item,fitness(item)) for item in population[:5]]
  fitness_list.sort(key=key)
  return fitness_list[0][0],fitness_list[1][0]
```

#Recombination

```python
import random
def recombination(gen1,gen2):
  x_point=random.randint(1,7)
  res1=res2=[]
  for i in range(x_point,7):
    x_point=i
    child1=gen1[:x_point] + gen2[x_point:]
    child2=gen2[:x_point] + gen1[x_point:]
    tmp1=[item[1] for item in child1]
    tmp2=[item[1] for item in child2]
    if len(tmp1) != len(set(tmp1)):
      res1=child1;
    if len(tmp2) != len(set(tmp2)):
      res2=child2
  if(res1==[]):
    res1=gen1
  if(res2==[]):
    res2=gen2
  return res1,res2
```

#Mutation - Swap

```python
def swap(gen):
  point1=random.randint(1,7)
  point2=random.randint(1,7)
  first=gen[point1]
  second=gen[point2]
  gen[point1]=second
  gen[point2]=first
  return gen
```

#Servival Selection

```python
def servival_selection(population,child1,child2):
  fitness_list=[(item,fitness(item)) for item in population]
  fitness_list.sort(key=key)
  #pop and push 2
  population=[item[0] for item in fitness_list]
  population[-1]=child1
  population[-2]=child2
  return population
```

#Run Algorithm

```python

population=init_population(100)
generation=0
p1=p2=c1=c2=[]
for generation in range(0,10000):
  p1,p2=parent_selection(population)
  c1,c2=recombination(p1,p2)
  pm=random.randint(0,10)
  if(pm<=8):
    c1=swap(c1)
  pm=random.randint(0,10)
  if(pm<=8):
    c2=swap(c2)
  print("\nGeneration",generation,"\nParents loss:",(fitness(p1),fitness(p2)),"\nChildren loss:",(fitness(c1),fitness(c2)))
  if fitness(c1)==0:
    print("Solution Found at Generation",generation,"\n",c1)
    break
  if fitness(c2)==0:
    print("Solution Found at Generation",generation,"\n",c2)
    break
  population=servival_selection(population=population,child1=c1,child2=c2)
for item in population:
  if(fitness(item)==0):
    print("res",item)
```
